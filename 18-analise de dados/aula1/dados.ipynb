```python
import pandas as pd 

# Data frame - df

df = pd.read_csv('estoque.csv')

# ver primeiras linhas do data frame
display(df.head())


print("\n____________________\n")

# ver as colunas do data frame
print(df.columns)


for i in df.columns:
    print(i)



```

```python
# ver uma coluna específica

print(df['marca'])

print("\n_______________________________________________________________________________\n")

# ver varias colunas 

print(df[['quantidade', 'modelo']])
```

```python
# agrupar os dados 
# Quantos tipos de cada estoque eu tenho?

Estoque_por_tipo = df.groupby('tipo')['tipo'].count()

print(Estoque_por_tipo)

print("\n_______________________________________________________________________________\n")

# Agrupar por cores 
Estoque_por_cores = df.groupby('cor')['cor'].count()

display(Estoque_por_cores)


```

```python
# Criar uma coluna nova

df['faturamento'] = df['quantidade'] * df['valor']
display(df)
```

```python
# Qual o valor total do estoque ?

valor_total = df['faturamento'].sum()

print(f'O valor total do estoque é de R$ {valor_total}')

estoque = df['quantidade'].sum()
print(f'A quantidade total do estoque é de {estoque} veículos')

# Qual a quantidade de veículos por tipo no estoque?

tipo_estoque = df.groupby('tipo')['quantidade'].sum()
print(tipo_estoque)


print("\n_______________________________________________________________________________\n")
# Qual a quantidade total de veículos por marca no estoque?

marca_valor = df.groupby('marca')['faturamento'].sum()
print(marca_valor)
```

```python
# Filtros 
#filtrar o estoque com veículos com até 15 anos de uso

novos = df[df['ano'] >= 2010]

display(novos)

# Filtrar veículos com valor igual ou acima de R$ 100.000,00
veículos_caros = df[df['valor'] >= 100000]

print("\n_______________________________________________________________________________\n")
#Carregar apenas carros 

carros = df[df['tipo']=='carro']
display(carros)

print("\n_______________________________________________________________________________\n")
# Carregar apenas motos

motos = df[df['tipo']=='moto']
display(motos)
print("\n_______________________________________________________________________________\n")

# Filtrar veículos apenas da cor marrom
veículos_marrons = df[df['cor']=='marrom']
display(veículos_marrons)

print("\n_______________________________________________________________________________\n")

# filtrar apenas caminhões
caminhões = df[df['tipo']=='caminhão']
display(caminhões)
```

```python
# Criar uma consulta de veículos pela cor

print("consulta de veículos pela cor:")
cor = input("Digite a cor desejada:")
consulta_cor = df[df['cor'] == cor]
display(consulta_cor)
```
