```python
# Quanto foi vendido para cada tipo de cartão de crédito ?
import pandas as pd 

df = pd.read_csv('Ecommerce_Purchases.csv')

cada_tipo_cartao = df.groupby('CC Provider')['Purchase Price'].sum()
print(cada_tipo_cartao)

```

```python
# Para quantos clientes diferentes essa loja vendeu ?

clientes_diferentes = df['Company'].unique()
resposta = len(clientes_diferentes)
print(f'vendemos para {resposta} clientes diferentes')
```

```python
# Em quantos idiomas foi vendido ?

idiomas = df['Language'].unique()
resposta_idiomas = len(idiomas)
print(f'Vendemos em {resposta_idiomas} idiomas ')
```

```python
# Qual tipo de trabalho mais comprou na loja ?

tipo_trabalho = df['Job'].max()

print(tipo_trabalho) 
```
