**Importe o pandas como pd.**

```python
import pandas as pd
```

**Leia o arquivo Salaries.csv como um DataFrame chamado "sal".**

```python
sal = pd.read_csv("Salaries.csv")
```

**Verifique o "head" do DataFrame.**

```python
display(sal.head())
```

**Use o método .info () para descobrir quantas entradas existem.**

```python
print(sal.info())

```

**Quantidade de funcionários e por JobTitle**<br>
**Total gasto e por agência**<br>
**Total de horas extras**<br>

```python
# Quantidade de funcionários 
func = sal['EmployeeName'].unique()
qtdefunc = len(func)
print(f'quantidade de funcionários: {qtdefunc}')


# Quantidade de funcinarios por cargo (job title)

jobfunc = sal.groupby('JobTitle') ['Id'].count()
print(jobfunc)


print("_________________________________________________________________________________")
# Total de horas extras pagas (overtime pay)

total_extras = sal['OvertimePay'].sum()
print(f'Total de horas extras pagas em $: {total_extras}')

```

**Qual é o "BasePay" médio?**

```python
mediasal = sal['BasePay'].mean()
print(f'Média de salário base anual em $: {mediasal:.2f}')
```

**Qual é a maior quantidade de "OvertimePay" no conjunto de dados?**

```python
maior_pagamento_em_hora_extra = sal['OvertimePay'].max()
print(f"Maior valor pago em horas extras : {maior_pagamento_em_hora_extra}" ) 
```

**Qual é o cargo de JOSEPH DRISCOLL? Nota: use todas as maiúsculas, caso contrário você pode obter uma resposta que não coincida (há também um Joseph Driscoll com minúsculas).**

```python
cargo_joseph = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']

display(cargo_joseph)
```

**Quanto JOSEPH DRISCOLL ganha (incluindo benefícios)?**

```python
Ganho_joseph = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits']

print(Ganho_joseph)
```

**Qual o nome da pessoa mais bem paga (incluindo benefícios)?**

```python
maxSal = sal['TotalPayBenefits'].max()
maxsalarioemp = sal[sal['TotalPayBenefits'] == maxSal]['EmployeeName'].values[0]
print(f'Nome da pessoa com maior salario é : {maxsalarioemp} com um salario de: $ {maxSal}')
```

**Qual o nome da pessoa paga mais baixa (incluindo benefícios)? Você percebe algo estranho sobre o quanto ele ou ela é paga?**

```python
lowSal = sal['TotalPayBenefits'].min()
lowsalarioemp = sal[sal['TotalPayBenefits'] == lowSal]['EmployeeName'].values[0]
print(f'Nome da pessoa com menor salario é : {lowsalarioemp} com um salario de: $ {lowSal}')
```

**Qual foi a média (média) BasePay de todos os funcionários por ano? (2011-2014)?**

```python
mediasalporano = sal.groupby('Year')['BasePay'].mean()
print(mediasalporano)

```

**Quantos títulos de trabalho únicos existem?**

```python
trabalhounicos = sal['JobTitle'].nunique()
print(trabalhounicos)
```

**Quais são os 5 principais empregos mais comuns?**

```python
cincoempregoscomum = sal['JobTitle'].value_counts().head(5)
print(cincoempregoscomum)
```

**Quantos Títulos de Trabalho foram representados por apenas uma pessoa em 2013? (Por exemplo, títulos de trabalho com apenas uma ocorrência em 2013?)**

```python
cincoempregoscomum = sal[sal['Year'] == 2013] 
onejob = sum(cincoempregoscomum['JobTitle'].value_counts() == 1)

print(onejob)
```
